```python
import pandas as pd

# Load CSV file
data = pd.read_csv('./source/pythonkurz/data.csv', header=None)
data
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(data.iloc[:,0], data.iloc[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN #conda install scikit-learn
from sklearn.cluster import KMeans
import numpy as np
import toolz


zipped = list(zip(data.iloc[:, 0], data.iloc[:, 1]))
result = DBSCAN(eps=50, min_samples=3).fit_predict(zipped)

with_result = list(zip(data.iloc[:, 0], data.iloc[:, 1], result))

grouped_data = toolz.groupby(lambda x: x[2], with_result)

for key, value in grouped_data.items():
    plt.scatter(list(map(lambda x: x[0] ,value)), list(map(lambda x: x[1] ,value)), label=key)


# plt.scatter(data.iloc[:, 0], data.iloc[:, 1], label=2)
```
